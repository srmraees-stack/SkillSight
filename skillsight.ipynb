Extraction Loop

```python
import pdfplumber
import pandas as pd
import os

# Define the path to your Drive folder
folder_path = '/content/drive/MyDrive/SkillSight_Data/'

# Create a list of your 3 PDF files
pdf_files = [f for f in os.listdir(folder_path) if f.endswith('.pdf')]

all_extracted_text = []

# Loop through each report
for file in pdf_files:
    file_path = os.path.join(folder_path, file)
    print(f"Extracting data from: {file}")

    with pdfplumber.open(file_path) as pdf:
        report_text = ""
        for page in pdf.pages:
            report_text += page.extract_text() + "\n"

        all_extracted_text.append({"Report_Name": file, "Raw_Text": report_text})

# Convert to a Pandas DataFrame for your model
df_skills = pd.DataFrame(all_extracted_text)
print("Extraction Complete! Data ready for SkillSight.")
```
